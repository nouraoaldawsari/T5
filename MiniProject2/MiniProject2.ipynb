## Creating a Pivot Table

In this section, we create a pivot table to restructure our data. This pivot table will have dates as rows and severity levels as columns, with the count of accidents as values. This format is useful for time series forecasting.

### Pivot Table Creation
We use the `pivot_table` method to create a pivot table, filling any missing dates with zeros. We then reset the index to ensure the date column is properly formatted.

```python
severity_pivot = severity_counts.pivot_table(index='Date', columns='Severity', values='Count', aggfunc='sum', fill_value=0)
severity_pivot = severity_pivot.asfreq('D', fill_value=0)
severity_pivot.reset_index(inplace=True)
severity_pivot
```

# Preprocessing

## Data Normalization

In this section, we normalize the data for severity level 2 accidents using the MinMaxScaler from the `sklearn.preprocessing` module. Normalization is an important step in preparing data for machine learning models, as it ensures that all features are on a similar scale.

### Applying MinMaxScaler
We use the `MinMaxScaler` to scale the data for severity level 2 accidents to a range between 0 and 1.


```python
from sklearn.preprocessing import MinMaxScaler

scale =  MinMaxScaler()
severity_pivot_2  = scale.fit_transform(severity_pivot_2)
```

## Creating Sequences for Time Series Forecasting

In this section, we prepare the data for time series forecasting by creating sequences. We use a sliding window approach to generate input-output pairs for our models.

### Sequence Creation Function
We define a function `create_sequences` that takes the normalized data and the number of timesteps (`n_steps`) as input. This function generates sequences of data to be used for training the models.

### Generating Sequences
We use the `create_sequences` function to generate the input (`X`) and output (`y`) sequences. In this example, we use the past 50 days to predict the next day's value.

### Checking the Shape of Sequences
We print the shapes of `X` and `y` to ensure they are correctly formatted for model training.


```python
import numpy as np
```

```python
def create_sequences(data, n_steps):
    X, y = [], []
    for i in range(len(data) - n_steps):
        X.append(data[i:i+n_steps])
        y.append(data[i+n_steps])
    return np.array(X), np.array(y)

# Number of timesteps (e.g., using the past 7 days to predict the next)
n_steps = 50

# # Convert the dataframe to a numpy array
# severity_data = severity_pivot.drop(columns='Date').values

# Create sequences
X, y = create_sequences(severity_pivot_2, n_steps)

# Check the shape of X and y
print(X.shape)  # Expected shape: (samples, timesteps, features)
print(y.shape)  # Expected shape: (samples, features)
```

```python
import matplotlib.pyplot as plt

# Plotting a few sequences with specified purple theme
plt.figure(figsize=(12, 6))
for i in range(3):  # Plotting the first 3 sequences
    plt.plot(X[i], label=f'Sequence {i+1}', color=['#2D3880', '#8382D1', '#9D9DE4'][i])  # Different purples for each sequence
plt.title('Sample Sequences for Time Series Forecasting', color='#615EDB')  # Title purple
plt.xlabel('Time Steps', color='#9D9DE4')  # Lighter purple
plt.ylabel('Normalized Number of Accidents', color='#9D9DE4')  # Lighter purple
plt.legend()
plt.grid(True, color='#CBCBEB', linestyle='--')  # Lightest purple for grid
plt.tight_layout()
plt.show()
```

# Models Building

## Building and Training the RNN Model

In this section, we define, compile, and train a Recurrent Neural Network (RNN) model using the `tensorflow.keras` library. We also implement early stopping to prevent overfitting.

### Model Definition
We define a Sequential model with multiple `SimpleRNN` layers. The model architecture includes:
- An input layer with 50 units and `return_sequences=True`
- Two hidden layers with 100 units each and `return_sequences=True`
- A final hidden layer with 50 units
- An output `Dense` layer with 1 unit

### Model Compilation
We compile the model using the `mean_squared_error` loss function and the `adam` optimizer.

### Early Stopping
We implement early stopping to monitor the validation loss and stop training if it doesn't improve for 10 consecutive epochs. This helps in restoring the best weights.

### Model Training
We train the model for 20 epochs with a batch size of 32 and a validation split of 20%.

### Predictions
After training, we use the model to make predictions on the input data.


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.callbacks import EarlyStopping

# Define the RNN model
model_rnn = Sequential()
model_rnn.add(SimpleRNN(50, input_shape=(X.shape[1], 1),  return_sequences=True))
model_rnn.add(SimpleRNN(100, return_sequences=True))
model_rnn.add(SimpleRNN(100, return_sequences=True))
model_rnn.add(SimpleRNN(50))

model_rnn.add(Dense(1))

model_rnn.compile(loss='mean_squared_error', optimizer='adam')

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

history_rnn = model_rnn.fit(X, y, epochs=20, batch_size=32, validation_split=0.2,  callbacks=[early_stopping])

predictions_rnn = model_rnn.predict(X)
```

## Building and Training the LSTM Model

In this section, we define, compile, and train a Long Short-Term Memory (LSTM) model using the `tensorflow.keras` library. LSTM networks are well-suited for time series forecasting due to their ability to capture long-term dependencies.

### Model Definition
We define a Sequential model with multiple `LSTM` layers. The model architecture includes:
- An input layer with 50 units and `return_sequences=True`
- Two hidden layers with 100 units each and `return_sequences=True`
- A final hidden layer with 50 units
- An output `Dense` layer with 1 unit

### Model Compilation
We compile the model using the `mean_squared_error` loss function and the `adam` optimizer, and we also track the `accuracy` metric.

### Model Training
We train the model for 20 epochs with a batch size of 32 and a validation split of 20%.

### Predictions
After training, we use the model to make predictions on the input data.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Define the RNN model
model_lstm = Sequential()
model_lstm.add(LSTM(50, input_shape=(X.shape[1], 1),  return_sequences=True))
model_lstm.add(LSTM(100, return_sequences=True))
model_lstm.add(LSTM(100, return_sequences=True))
model_lstm.add(LSTM(50))

model_lstm.add(Dense(1))

model_lstm.compile(loss='mean_squared_error',optimizer='adam', metrics=['accuracy'])
# Train the model
history_lstm = model_lstm.fit(X, y, epochs=20, batch_size=32, validation_split=0.2)

# Make predictions
predictions_lstm = model_lstm.predict(X)
```

## Building and Training the GRU Model

In this section, we define, compile, and train a Gated Recurrent Unit (GRU) model using the `tensorflow.keras` library. GRU networks are well-suited for time series forecasting due to their ability to capture long-term dependencies with a simpler architecture compared to LSTM.

### Model Definition
We define a Sequential model with multiple `GRU` layers. The model architecture includes:
- An input layer with 50 units and `return_sequences=True`
- Two hidden layers with 100 units each and `return_sequences=True`
- A final hidden layer with 50 units
- An output `Dense` layer with 1 unit

### Model Compilation
We compile the model using the `mean_squared_error` loss function and the `adam` optimizer, and we also track the `accuracy` metric.

### Model Training
We train the model for 20 epochs with a batch size of 32 and a validation split of 20%.

### Predictions
After training, we use the model to make predictions on the input data.


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense

# Define the RNN model
model_gru = Sequential()
model_gru.add(GRU(50, input_shape=(X.shape[1], 1),  return_sequences=True))
model_gru.add(GRU(100, return_sequences=True))
model_gru.add(GRU(100, return_sequences=True))
model_gru.add(GRU(50))

model_gru.add(Dense(1))

model_gru.compile(loss='mean_squared_error',optimizer='adam', metrics=['accuracy'])
# Train the model
history_gru = model_gru.fit(X, y, epochs=20, batch_size=32, validation_split=0.2)

# Make predictions
predictions_gru = model_gru.predict(X)
```

### Evaluation Results
